<b>split the input of words into characters</b>

```python
def split_words(inputs):
    char_list=[]
    for ch in inputs:
        char_list.append(ch)
    return char_list
```

```python
split_words('abc#123')
```

<b>Let's use TF-IDF vectorizer to convert String data into numerical data</b>

```python
vectorizer=TfidfVectorizer(tokenizer=split_words)
```

<b>apply TF-IDF vectorizer on data</b>

```python
p=vectorizer.fit_transform(key1)
```

```python
p.shape
```

```python
vectorizer.get_feature_names()
```

```python
first_vec=p[0]
first_vec
```

```python
first_vec.T.todense()
```

```python
dataset=pd.DataFrame(first_vec.T.todense(),index=vectorizer.get_feature_names(),columns=['TF-IDF'])
dataset.sort_values(by=['TF-IDF'],ascending=False)
```

<b>split data into train & test</b>
<br>train---> To learn the relationship within data,<br>
test-->  To do predictions, and this testing data will be unseen to my model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(p,key2,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

<b>Apply Logistic on data as use-cas is Classification</b>

```python
clf=LogisticRegression(random_state=0,multi_class='multinomial')
```

```python
clf.fit(X_train,y_train)
```

<b>prediction for specific custom data</b>

```python
dpt=np.array(['%@123abcd'])
pred=vectorizer.transform(dpt)
clf.predict(pred)
```

```python
y_pred=clf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
